```python
from pathlib import Path
import pandas as pd

data_dir = Path("./mgat1ko_data/nav")

cm = pd.read_excel(data_dir/"capacitance.xlsx")
```

## Normalize current traces by the cell capacitance

```python
fpaths = (data_dir/"wt-trace").glob("*.xlsx")
fnames = [f.stem for f in fpaths]

for idx,row in cm[cm['group']=='WT'].iterrows():
    if str(row.file_name) in fnames:
        print(f"File {idx}: {row.file_name} is in file list")
    else:
        print(f"File {idx}: {row.file_name} is NOT in file list")
```

```python
fpaths = (data_dir/"mgat1ko-trace").glob("*.xlsx")
fnames = [f.stem for f in fpaths]

for idx,row in cm[cm['group']=='MGAT1KO'].iterrows():
    if str(row.file_name) in fnames:
        print(f"File {idx}: {row.file_name} is in file list")
    else:
        print(f"File {idx}: {row.file_name} is NOT in file list")
```

```python
for idx,row in cm[cm['group']=='WT'].iterrows():
    p = data_dir/"wt-trace"/(str(row.file_name)+".xlsx")
    ctrace = pd.read_excel(p)
    ctrace.iloc[:,2:] = ctrace.iloc[:,2:]/long(row.cm)
```
